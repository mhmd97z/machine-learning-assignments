## Loading dataset
Load the IMDB-Movie-Data.csv file into a pandas dataframe and show its first 5 rows. Each row of this dataframe corresponds to one film.

```python
dataset_df = pd.read_csv('IMDB-Movie-Data.csv')
dataset_df.head()
```

## Calculating Profit
Add a column to the dataframe that calculate the profit of the movies. (Suppose the inflation rate is 3% and 1990 as the base year)

```python
dataset_df['Profit'] = dataset_df.apply(lambda x: x['Revenue (Millions)'] / (1.03 ** (x['Year'] - 1990)), axis=1)
dataset_df
```

## Nolan Movies
Filter Christofer Nolan movies and sort them in a descending order with respect to their profits then show the result.

```python
nolan_df = dataset_df[dataset_df['Director'] == 'Christopher Nolan'].sort_values(by='Profit',ascending=False)
nolan_df
```

## Favorite Genres
Calculate the revenues of different genres in a dataframe and sort them in a descending order. Then show the result.

```python
genre_df = dataset_df[['Genre', 'Revenue (Millions)']]
genre_df['Genre'] = genre_df['Genre'].str.split(',')

genre_df = (genre_df
 .set_index(['Revenue (Millions)'])['Genre']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_1', axis=1)
 .rename(columns={0:'Genre'}))

genre_df = genre_df.groupby('Genre').sum().sort_values(by='Revenue (Millions)',ascending=False)

genre_df
```

## Actors/Actresses Activity
For each Actors/Actresses caculate the number of movies in the dataframe, then sort the actors/actresses first by their number of movies in descending order and then by their name in ascending order. 

```python
actors_df = pd.DataFrame(pd.DataFrame(dataset_df.Actors.str.split(', ').to_list()).stack())
actors_df.columns = ['actor_name']
actors_df['number_of_movies'] = 1
actors_df = actors_df.groupby('actor_name').sum()
print(actors_df.sort_values(by='number_of_movies',ascending=False))
print(actors_df.sort_values(by='actor_name'))
```
