## ECDF and Functions 

Write a function that takes a one-dimensional NumPy array as input and returns two arrays of x and y which together are going to draw an empirical distribution function or ECDF. 

```python
def ecdf(data):
    min_val = data.min()
    max_val = data.max()
    step_size = 1 / len(data)
    x = np.arange(min_val, max_val + step_size, step_size)
    y = np.zeros(x.shape)
    for i in range(len(x)):
        y[i] = len(data[data < x[i]]) / len(data)
    return [x, y]
```
