In this section, you need to prepare your data. One of the problems you may face in datasets are missing values. There are various strategies to treat missing values. First, you can delete any sample with missing values. Another method is to fill missing values with mean of the feature. 

```python
idx, idy = np.where(pd.isnull(Data_set))
result = np.column_stack([Data_set.index[idx], Data_set.columns[idy]])
# delete any sample with a missing value 
clear_data_set = Data_set.drop(Data_set.index[idx])
X1= clear_data_set.iloc[:,0:13]
y1= clear_data_set.iloc[:,13]
# replace missing values with mean of the features for both X and y 
col_mean = clear_data_set.mean()
temp = Data_set
for elm in result:
    if elm[1] == 'quality':
        temp[elm[1]][elm[0]] = temp['quality'].mode()
    else:
        temp[elm[1]][elm[0]] = col_mean[elm[1]]
    
X2= temp.iloc[:,0:13]
y2= temp.iloc[:,13]
```

Implement the Logistic regression algorithm. You can change everything in the cell below.

```python
class LogisticRegression:
    def __init__(self, lr=0.1, num_iter=100000, fit_intercept=True, verbose=False): ## if fit_intercept==True you should increase the Input dimension by 1
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.verbose = verbose
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def probability(theta, x):
        return sigmoid(np.dot(x, theta))

    def cost_function(self, theta, x, y):
        m = x.shape[0]
        return = -(1/m)*np.sum(y*np.log(probability(theta, x)) + (1 - y)*np.log(1 - probability(theta, x)))

    def fit (self, X, y) : 
        weight=np.zeros(X.shape[1]+1)
        X_train=(np.hstack((np.ones((X.shape[0],1)),X)))
        loss_per_iteration = np.zeros(self.num_iter)
        missmatch = np.zeros(self.num_iter)
        
        for i in range(self.num_iter):
            
            sig_input = np.dot(weight,X_train.T)
            y_prime = self.predict_prob(sig_input)
            update_value = np.dot((y_prime-y),X_train)
            weight = weight-(self.lr/X.shape[0])*update_value
            loss_per_iteration[i] =  -(1/len(y))*(np.dot(y,np.log(y_prime))+np.dot(1-y,np.log(1-y_prime)))
            y_predict = [1 if l>0.5 else -1 for l in y_prime];
            missmatch[i] = np.shape(np.where(y_predict!=y))[1]
    
        return weight,loss_per_iteration,missmatch   
    
    def predict_prob (self, X) :
        return 1/(1+np.exp(-X))
      
    def predict (self, X, w) :
        X_train = (np.hstack((np.ones((X.shape[0],1)),X)))
        y_value = self.predict_prob(np.dot(w,X_train.T))
        y_prime = [1 if l>0.5 else -1 for l in y_value]
        return y_prime
```

Note that you should plot the dataset alongside the seprator line generated by Logistic regression using the first two features in both dataset generated after missing values treatment and for different learning rates. Also report accuracy for classification using all features for different learning rates, lr= {0.01, 0.1, 1, 10}.

```python
predict_line_x = np.linspace(np.min(X['fixed acidity']),np.max(X['volatile acidity']),30)
lrs = [0.01, 0.1, 1, 10]
# First method: delete any sample with a missing value
iter_num = 10000
print(y1.unique().tolist())
data_with_outlier = X1.iloc[:,1:].to_numpy()
label_with_outlier = y1.to_numpy()
rem_indices = (np.abs(stats.zscore(X1)) < 3).all(axis=1) 
data_without_outlier = X1[rem_indices]
label_without_outlier = y1[rem_indices]
for lr_ in lrs:
    model = LogisticRegression(lr=lr_,num_iter=iter_num)
    #part1: with outlier data
    w,loss_per_iteration,missmatch  = model.fit(data_with_outlier,label_with_outlier)
    y_predict=model.predict(data_with_outlier,w)
    plt.figure()
    plt.title("learning rate= "+str(lr_) + "For the 1st method of removing NANs")
    plt.scatter(data_with_outlier[:,0], data_with_outlier[:,1],color=["blue" if l==1 else "orange" for l in y_predict])
    plt.plot(predict_line_x,(-w[1]/w[2])*predict_line_x + (-w[0]/w[2]),'--g')
    plt.xlabel('X0')
    plt.ylabel('X1')
    plt.ylim((np.min(data_with_outlier[:,1])-1,np.max(data_with_outlier[:,1])+1))
    plt.show()
```
