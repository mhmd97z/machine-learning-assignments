*I checked my class with different variables and it seems a learning rate about 0.02 is optimal and my final result is:*

````python
class Optimizer:
    def __init__(self, lr=0.02 ,weight=[5,-5,5,-5], num_iter=6000): 
        self.lr = lr
        self.num_iter = num_iter
        self.weight = weight
        
    def MSE(self,X,y) :
        
        weight=self.weight
        X_train=(np.hstack((X,np.ones((X.shape[0],1)))))

        for i in range(self.num_iter):
            y_prime=np.dot(weight,X_train.T)
            weight=weight-self.lr*(np.dot((y_prime-y),X_train)/y.shape[0])
        loss_iter=np.sum(np.power((y-y_prime),2))/y.shape[0]  
        return weight,loss_iter,(y_prime-y)


model = Optimizer()
weight, loss_iter, A = model.MSE(x[:300],y[:300])


print('MSE is '+str(np.round(loss_iter,4)))
print('optimal [Weghits,Bias] is '+str(weight))
print('learning rate is '+ str(0.02))
````

*Now I changed my class to a online optimazer with outlier-removed dataset, first 300 rows are for initial fitting and the other 70 rows are time series like.*

*We have two kinds of optimizer (static learning rate and dynamic learning rate):* 


1.   static learning rate = 0.025

  ```

2.   dynamic learning rate = 0.025

  ```


**also, the largest posibble learning rate for convergence is about 0.036.**

````python
class OnlineOptimizer:
    def __init__(self, lr=0.025 ,weight=[5,-5,5,-5], num_iter=6000): 
        self.lr = lr
        self.num_iter = num_iter
        self.weight = weight
        self.finalWeight = weight
        
    def InitialMSE(self,X,y):

        weight=self.weight
        X_train=(np.hstack((X,np.ones((X.shape[0],1)))))

        for i in range(self.num_iter):
            y_prime=np.dot(weight,X_train.T)
            weight=weight-self.lr*(np.dot((y_prime-y),X_train)/y.shape[0])
        loss_iter=np.sum(np.power((y-y_prime),2))/y.shape[0] 
        self.initial_weight = weight

    def MSE(self,X,y,X_online,y_online,time,onlineRate=0,terminated=0) :
        if (time==1):
          weight_online = self.initial_weight
        else :
          weight_online = self.weight

        X_train=(np.hstack((X_online,np.ones((X_online.shape[0],1)))))
        y_prime=np.dot(weight_online,X_train.T)
        
        lr = self.lr
        if (onlineRate==1):
          lr = self.lr/np.sqrt(time)
        
        self.weight=weight_online-lr*(np.dot((y_prime-y_online),X_train)/y_online.shape[0])

        if (time==1):
          self.finalWeight = self.weight
        else :
          self.finalWeight = self.finalWeight + weight_online

        X_train=(np.hstack((X,np.ones((X.shape[0],1)))))
        y_mse=np.dot(self.finalWeight/time,X_train.T)
        loss_iter=np.sum(np.power((y-y_prime),2))/y.shape[0]

        y_mse=np.dot(self.finalWeight/time,X_train.T)
        loss_iter=np.sum(np.power((y-y_mse),2))/y.shape[0]

        if (terminated==0):
          return loss_iter
        if (terminated==1):
          return self.finalWeight/time,loss_iter

model = OnlineOptimizer()
model.InitialMSE(x[:300],y[:300])

index = 300
loss_iter = np.zeros(70)
for i in range(70):
  if (i<69):
    loss_iter[i] = model.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=0)
  else :
    weight,loss_iter[i]=model.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=0,terminated=1)

plt.figure()
plt.plot(range(70),loss_iter)

model2 = OnlineOptimizer()
model2.InitialMSE(x[:300],y[:300])

index = 300
loss_iter2 = np.zeros(70)
for i in range(70):
  if (i<69):
    loss_iter2[i] = model2.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=1)
  else :
    weight2,loss_iter2[i]=model2.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=1,terminated=1)

plt.plot(range(70),loss_iter2)
plt.xlabel('Time')
plt.ylabel('MSE(t)')
plt.title('Learning Rate=0.025')
plt.show()

print('MSE(70) for first method is '+str(np.round(loss_iter[69],4))+' and '+ 'MSE(70) for second method is '+ str(np.round(loss_iter2[69],4)))
print('[Weghits,Bias] for first method is '+str(weight))
print('[Weghits,Bias] for second method is '+ str(weight2))


model = OnlineOptimizer(lr=0.036)
model.InitialMSE(x[:300],y[:300])

index = 300
loss_iter = np.zeros(70)
for i in range(70):
  if (i<69):
    loss_iter[i] = model.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=0)
  else :
    weight,loss_iter[i]=model.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=0,terminated=1)

plt.figure()
plt.plot(range(70),loss_iter)

model2 = OnlineOptimizer(lr=0.036)
model2.InitialMSE(x[:300],y[:300])

index = 300
loss_iter2 = np.zeros(70)
for i in range(70):
  if (i<69):
    loss_iter2[i] = model2.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=1)
  else :
    weight2,loss_iter2[i]=model2.MSE(x[:index+i],y[:index+i],x.iloc[[index+i]],y.iloc[[index+i]],i+1,onlineRate=1,terminated=1)

plt.plot(range(70),loss_iter2)
plt.title('MSE(t) for Two Methods With Learning Rate=0.036')
plt.xlabel('Time')
plt.ylabel('MSE(t)')

plt.show()
````
